<font color='#2a9d8f' size=5>Evaluating Conditions with Comparison and Logical Operators</font>

Comparison and logical operators are essential tools in Python, allowing us to make decisions based on data. This is especially useful in healthcare, for instance, when determining if a patient's measurements fall within a certain range or if multiple conditions are met for a diagnosis.

**Comparison Operators**–These operators are used to compare two values:

1. `==:` Equal to
2. `!=:` Not equal to
3. `<:` Less than
4. `>:` Greater than
5. `<=:` Less than or equal to
6. `>=:` Greater than or equal to

````python
5 > 6
````

````python
patient_temperature = 38.5

# Check if the temperature is above normal (37.5°C is considered average human temperature)
is_fever = patient_temperature > 37.5
print(is_fever)  # This will print True since 38.5 > 37.5

# Check if the temperature is exactly 38.5°C
is_specific_temp = patient_temperature == 38.5
print(is_specific_temp)  # This will print True
````

**Logical Operators**–These operators are used to combine conditional statements:

1. `and`: Returns True if both statements are true
2. `or`: Returns True if one of the statements is true
3. `not`: Reverse the result, returns False if the result is true and vice versa

````python
patient_temperature = 38.5
patient_heart_rate = 110

# Check if the patient has fever and an elevated heart rate (normal resting heart rate is 60-100 beats per minute for adults)
is_concerning = (patient_temperature > 37.5) and (patient_heart_rate > 100)
print(is_concerning)  # This will print True since both conditions are met

# Check if the patient has either fever or an elevated heart rate
either_condition = patient_temperature > 37.5 or patient_heart_rate < 50
print(either_condition)  # This will print True since the fever condition is met

# Check if the patient's temperature is NOT within the normal range (36.1°C to 37.2°C)
is_not_normal = not(36.1 <= patient_temperature <= 37.2)
print(is_not_normal)  # This will print True since 38.5 is outside the normal range
````

<font color='#2a9d8f' size=5> Exploring Tuples: The Immutable Collections in Python </font>

A tuple, in Python, acts as an ordered collection of items. Tuples are akin to lists, but with a key difference: *they are immutable*. That means once they are created, their content cannot be altered. This immutability offers reliability, ensuring that data remains consistent.

**Creating and Displaying Tuples–**To construct a tuple, you will enclose your comma-separated values within parentheses `()`.

````python
medical_entry = ('John Doe', 1985, 'AB+', 'Hypertension')
print(medical_entry)
# Output: ('John Doe', 1985, 'AB+', 'Hypertension')
````

**Why Immutable?** Imagine a scenario where you're processing critical health data. If a piece of software accidentally changes a patient's blood type from `AB+` to `O+`, the repercussions could be severe. With tuples, such concerns are mitigated due to their immutable nature.

**Accessing Elements: Indexing and Slicing–**Much like strings and lists, tuple items can be accessed via their index. Remember, Python's indexing starts at **zero**.

````python
patient_name = medical_entry[0]
birth_year = medical_entry[1]
print(patient_name, "was born in", birth_year)
# Output: John Doe was born in 1985
````

**Probing Tuple Dimensions–**To discern the number of items nested within a tuple, employ the `len()` function.

````python

entry_length = len(medical_entry)
print(f"The tuple has {entry_length} items.")
# Output: The tuple has 4 items.
````

**Complex Structures: Nested Tuples–**For intricate data structures, tuples can be embedded within other tuples.

````python
address = ('221B Baker St', 'London', 'UK')
detailed_entry = ('Sherlock Holmes', 1854, address)
print(f"{detailed_entry[0]} resides at {detailed_entry[2][0]}, {detailed_entry[2][1]}.")
# Output: Sherlock Holmes resides at 221B Baker St, London.
````

**Unboxing Tuples–**When aware of a tuple's architecture, you can allocate its values to individual variables. This is termed as tuple unpacking.

````python
first_name, birth_year, blood_type, ailment = medical_entry
print(f"{first_name} suffers from {ailment}.")
# Output: John Doe suffers from Hypertension.
````

<font color='#2a9d8f' size=4> When to use Tuples? </font>
Given their **static** nature, tuples are primarily utilized when:
* Data should remain static throughout the program.
* Accidental data modification can have severe implications.

For instance, within health informatics, genetic sequences, which should not be altered once recorded, could be represented as tuples to prevent inadvertent mutations.

<font color='#2a9d8f' size=5> Introduction to Python Lists</font>

**What is a List?** In Python, a list is like a container where you can store multiple items. Think of it as a shelf, where you can place different things and arrange them in a specific order.

````python
patient_record = ['John Doe', 28, 'Fever']
print(patient_record)
# Output: ['John Doe', 28, 'Fever']
````

**Updating Health Records–**You can add new symptoms or details to a patient's record:

````python
patient_record.append('Cough')  # Add a symptom to the end
patient_record.insert(2, 'Male')  # Specify gender after age
print(patient_record)
# Output: ['John Doe', 28, 'Male', 'Fever', 'Cough']
````

If a symptom is no longer relevant or was recorded by mistake:

````python
patient_record.remove('Fever')
print(patient_record)
# Output: ['John Doe', 28, 'Male', 'Cough']
````

**Checking a Portion of the Record–**Slicing allows you to access specific subsets of a patient's record. By understanding how slicing works, you can quickly extract the information you need.

* With **positive** indices, you start counting from the beginning of the list.
* **Negative** indices allow you to count from the end of the list, which is helpful when you want to quickly check recent entries.

````python
patient_record = ['John Doe', 29, 'Male', 'Fever', 'Cough', 'Headache']

# Checking the name and age:
basic_info = patient_record[0:4] #get the items from index 0 up to (but not including) index 4
print(basic_info)
# Output: ['John Doe', 29, 'Male', 'Fever']

# Retrieving all symptoms:
symptoms = patient_record[3:]
print(symptoms)
# Output: ['Fever', 'Cough', 'Headache']

# Retrieving the first two symptoms:
first_two_symptoms = patient_record[3:5]
print(first_two_symptoms)
# Output: ['Fever', 'Cough']

# Getting the last symptom:
last_symptom = patient_record[-1]
print(last_symptom)
# Output: 'Headache'

# Getting the last two symptoms:
recent_symptoms = patient_record[-2:]
print(recent_symptoms)
# Output: ['Cough', 'Headache']

# Retrieving details excluding the last symptom:
record_without_last_symptom = patient_record[:-1]
print(record_without_last_symptom)
# Output: ['John Doe', 29, 'Male', 'Fever', 'Cough']
````

**Modifying Existing Information–**If you need to update the age after a birthday:

````python
patient_record[1] = 35
print(patient_record)
# Output: ['John Doe', 35, 'Male', 'Fever', 'Cough', 'Headache']
````

**Example [BMI (Body Mass Index) Classification]–**BMI is a useful index to categorize one's weight relative to their height. Let's create a program that helps classify an individual's BMI:

* In Python, the double asterisk `**` is used as the exponentiation operator. It is used to raise a number to a power, making it very useful for performing mathematical calculations involving exponentiation.
* The Python `input()` function serves the purpose of gathering user input from the keyboard while a program is running. When the `input()` function is invoked, the program momentarily halts its execution, anticipating the user to enter text and subsequently hit the **Enter** key.

````python
# Prompt the user to enter their weight in kilograms and height in meters
weight = float(input('Enter your weight in kg: '))
height = float(input('Enter your height in meters: '))

# Calculate the BMI (Body Mass Index) using the entered weight and height
bmi = weight / (height ** 2)
print(bmi)

# Determine the BMI category and provide a corresponding message
if bmi < 18.5:
    print('Underweight')  # Displayed when BMI is less than 18.5
elif 18.5 <= bmi < 24.9:
    print('Normal weight')  # Displayed when BMI is between 18.5 (inclusive) and 24.9 (exclusive)
elif 24.9 <= bmi < 29.9:
    print('Overweight')  # Displayed when BMI is between 24.9 (inclusive) and 29.9 (exclusive)
else:
    print('Obesity')  # Displayed when BMI is 29.9 or higher
````

**Example [Counting Above Average Heart Rates]–**Suppose we have a list of heart rates from a group of patients. Let's find out how many are above the average:


````python
heart_rates = [72, 75, 78, 80, 65, 74, 77, 83, 90, 85]

avg_rate = sum(heart_rates) / len(heart_rates)
above_avg_count = 0

for rate in heart_rates:
    print(rate)
    if rate > avg_rate:
        above_avg_count += 1

print(f'{above_avg_count} heart rates are above the average of {avg_rate}.')
````

**Mapping Diseases to Symptoms–**In health informatics, linking diseases to their potential symptoms can be essential. Using nested loops can be handy for such tasks:

````python
# Define a dictionary of diseases with associated symptoms
diseases = {
    'flu': ['fever', 'headache', 'cough'],
    'chickenpox': ['fever', 'spots', 'itchiness']
}

# Iterate through the diseases and their symptoms
for disease, symptoms in diseases.items():
    # Print the name of the disease
    print(f"\nSymptoms of {disease}:")

    # Iterate through the symptoms of the current disease
    for symptom in symptoms:
        # Print each symptom with a hyphen
        print(f"- {symptom}")
````

**Tracking Patient Visits–**Consider a scenario where a health facility wants to track the frequency of patient visits. A dictionary can be an ideal structure for this:

````python
patients = {
    'Alice': ['01-Jan', '15-Feb', '10-Mar'],
    'Bob': ['05-Jan', '20-Mar']
}

for patient, visits in patients.items():
    print(f"\n{patient} visited on:")
    for visit in visits:
        print(f"- {visit}")
````

You can use range() function to iterate on a range of numbers

````python
# Define a vaccination schedule as a dictionary with ages as keys and corresponding vaccines as values
vaccination_schedule = {
    0: "Hepatitis B",
    1: "DTP, Hib, Polio, Pneumococcal, Rotavirus",
    6: "Influenza (Yearly)",
    12: "HPV, Meningococcal ACWY",
    18: "Meningococcal B"
}

# Print the title of the vaccination schedule
print("Vaccination Schedule from ages 0 to 18:")

# Iterate through ages from 0 to 18
for age in range(0, 19):
    print(age)
    #Check if the current age has a corresponding vaccine in the schedule
    if age in vaccination_schedule:
        # If there is a vaccine, print the age and the corresponding vaccine
        print(f"At age {age}, administer: {vaccination_schedule[age]}")
    else:
        # If there is no specific vaccine for the age, print that there are no specific vaccinations
        print(f"At age {age}, no specific vaccinations.")
````

`range()` generates a sequence of numbers for iteration, where you specify the start, end (*exclusive*), and step values, allowing precise control over loop behavior. When the above code runs, it will print out the vaccination schedule from ages `0` to `18`. The `range(19)` allows the loop to iterate from `0` up to, but not including, `19`. In the following code:

the `range(6, 20, 5)` allows the loop to iterate from age 6 up to, but not including, 20, in steps of 5 years.


In Python, a while loop serves as a programming construct that permits the repeated execution of a code block, ensuring that the execution continues as long as a particular condition remains valid. This fundamental mechanism is ideal for crafting loops that adapt to changing conditions, making it applicable to situations where **the number of iterations needed is uncertain**.

**Example [Monitoring Blood Pressure]–**Let's say a health app prompts users to input their blood pressure readings until they enter '**stop**'. The program will then inform them if any of the readings were above the normal range.

````python
# Prompt the user to enter blood pressure readings and provide instructions
print("Enter your blood pressure readings. Type 'stop' when done.")

# Initialize a variable to count high readings
high_readings = 0

# Start an infinite loop to receive readings until the user types 'stop'
while True:
    # Receive the blood pressure reading input from the user
    reading = input("Enter systolic/diastolic reading (e.g., 120/80): ")

    # Check if the user typed 'stop' to exit the loop
    if reading == 'stop':
        break

    # Split the input into systolic and diastolic values and convert them to integers
    systolic, diastolic = map(int, reading.split('/'))

    # Check if either the systolic or diastolic reading is above the normal range
    if systolic > 120 or diastolic > 80:
        high_readings += 1

# Check if there were any high blood pressure readings
if high_readings:
    print(f"You had {high_readings} high blood pressure readings.")
else:
    print("All your readings were normal.")
````

**Vaccine Distribution–**Imagine a scenario where a clinic distributes vaccines to citizens. Due to limited availability, the clinic can only vaccinate a certain number of people daily. Here, we use the while loop to simulate the process.

````python
# Initialize the number of available vaccines for the day
vaccines_available = 100

# Start a loop to distribute vaccines while there are still some available
while vaccines_available > 0:
    # Print the number of vaccines remaining for the day
    print(f"\n{vaccines_available} vaccines left for the day.")

    # Prompt the user to enter the number of vaccines required for the next group
    num_vaccines_required = int(input("Enter the number of vaccines required for the next group: "))

    # Check if there are enough vaccines available for the next group
    if num_vaccines_required <= vaccines_available:
        # Reduce the available vaccine count and report the number of people vaccinated
        vaccines_available -= num_vaccines_required
        print(f"{num_vaccines_required} people vaccinated!")
    else:
        # Notify that there are not enough vaccines for the entire group
        print("Not enough vaccines for the entire group. Please wait for the next shipment.")

# Print a message when no more vaccines are left for the day
print("\nNo more vaccines left for the day. Please come back tomorrow!")
````

Imagine you are a public health expert working on a vaccination campaign. For each patient you encounter, there is a sequence of steps: assess the patient's health history, decide on the appropriate vaccine, and then administer it. Instead of considering all the individual decisions and actions every time, you think of the entire process as *vaccinate a patient*.

In the Python programming world, this concept translates to **functions**. A function is a self-contained block of code that encapsulates a specific task or related group of tasks. When working with health data, for instance, you might have a function that calculates a patient's *Body Mass Index (BMI)* or another that predicts potential health risks based on age and lifestyle. By calling these functions with the necessary data (or in medical terms, the "patient's information"), you can quickly obtain the results without diving into the intricate details each time.

Here is how we define the syntax of functions in Python:



```python
def function_name(parameters):
    # Function body (code block)
    # Perform actions here
    return result  # Optional
```

* `def`: This keyword signifies the start of a function definition. It tells Python that you are creating a reusable piece of code.
* `function_name`: Choose a descriptive name for your function. This name should reflect the purpose or action that the function performs.
* `parameters`: These are optional inputs that the function can accept. Think of parameters as labeled containers placed within the parentheses of your function definition. These containers are like open slots waiting to be filled with specific values when you decide to use the function.
* `return`: You can use the `return` statement to send a result back from the function *(optional)*. While not always required, it is incredibly useful when you want to capture and use the result of a function's computation elsewhere in your program.

**Example–**Let's design a simple function that calculates a patient's Body Mass Index (BMI). The formula for BMI is:

<br>

$$\text{BMI} = \frac{\text{weight (kg)}}{\text{height (m)}^2} $$

````python
# Defining a function to calculate BMI
def calculate_bmi(weight, height):
    bmi = weight / (height**2)  # Calculate BMI using the provided weight and height
    return bmi  # Return the calculated BMI value

# Using the function to get the BMI of a patient weighing 70kg and 1.75m tall
patient_bmi = calculate_bmi(70, 1.75)  # Calling the function with specific weight and height values
print(f"Patient's BMI: {patient_bmi:.2f}")  # Printing the calculated BMI with two decimal places
````

**Making Functions More Dynamic–**Functions can be made flexible to handle a variety of cases. Let's enhance our BMI function to also provide an assessment based on the BMI value.

````python

def bmi_assessment(weight, height):
    bmi = weight / (height**2)

    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 24.9 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"

# Let's use our function to get the assessment of a patient
assessment = bmi_assessment(70, 1.75)


print(f"Patient's assessment: {assessment}")
````

**Default Argument Values–**In some cases, a function may need a common default value.

Example: When collecting patient information, perhaps the majority of your patients are non-smokers, so we can have a default argument value as '*Non-smoker*'.

````python
def patient_info(name, age, smoking_status="Non-smoker"):
    return f"{name}, {age} years old, {smoking_status}"

print(patient_info("John Doe", 30)) # Will default to "Non-smoker"
print(patient_info("Jane Doe", 29, "Smoker"))
````
